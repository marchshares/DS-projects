#### Task 3.
Перед вами шахматная доска размером NxN. В верхнем левом углу находится шашка. Шашка может
двигаться только вправо или вниз (вверх или влево двигаться не может). Сколькими различными путями
шашка может прийти в нижний правый угол?


**Ответ**

Чтобы шашке дойти из левого верхнего угла в правый нижний, ей надо сделать N-1 шагов вправо и N-1 шагов вниз, причем в любом порядке. Каждый такой путь должен входить в нам результат. Т.е. шашке надо сделать (N-1)+(N-1) действий, причем половина из них - вниз, половина вправо.
Пронумеруем наши ходы от 1 до 2\*(N-1)
Соответственно, мы можем проставить сначала каких (N-1) действий из 2\*(N-1) были ходами направо, а оставшиеся будут ходами вниз. Количество всех таких сочетаний и будет нашим ответом.
$$ C_{2*(N-1)}^{N-1}={\frac {(2*(N-1))!}{(N-1)!\left((2*(N-1))-(N-1)\right)!}}$$

#### Task 4.
Осьминог Пауль 16 раз пробовал угадать победителя футбольного матча. 14 раз он угадал правильно, 2
раза ошибся.
Проверьте гипотезу H0: Пауль выбирает победителя футбольного матча случайно.
Сделайте вывод.


**Ответ**

Гипотеза H0: выбор случаен,т.е. не существует взаимосвязи между выбором осьминога и победителем матча => мы ожидаем, что из 16 раз осминог должен был угадать 8 и ошибться 8 раз
Альтернативная гипотеза: выбор не случаен
Наблюдаемые результаты: из 16 раз 14 раз правильно, 2 раза неправильно

Для проверки статистической значимости гипотезы H0 посчитаем p-value.
Выберем уровень значимости в p=0.01, т.е. если pvalue > 0.01, то это значит, что мы не можем отвегнуть H0, так как в таком случае у нас будет более чем 1% уверенность, что наблюдаемые результат был обусловлен случайностью

Для оценки pvalue посчитаем хи2:
хи2=(14-8)^2/8+(2-8)^2/8=9

По таблице когда хи2 > 6.635, то pvalue < 0.01. Значит получившееся pvalue меньше чем наш уровень значимости и мы можем отбросить нулевую гипотезу => резултат не случаен
